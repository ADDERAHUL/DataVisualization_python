```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
file_path = '/content/drive/MyDrive/samplesuper/Sample - Superstore.xlsx'
df = pd.read_excel(file_path)

# Display the first few rows of the dataset
df.head()
```

```python
# Group by state and calculate total sales
state_sales = df.groupby('State')['Sales'].sum().sort_values(ascending=False)

# Plotting state-wise sales
plt.figure(figsize=(12, 8))
state_sales.plot(kind='bar', color='skyblue')
plt.title('State-wise Sales')
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.show()
```

```python
# Filter for a specific state (e.g., 'Florida') and calculate city-wise sales
state = 'Florida'
city_sales = df[df['State'] == state].groupby('City')['Sales'].sum().sort_values(ascending=False)

# Plotting city-wise sales in the selected state as an area chart
plt.figure(figsize=(12, 8))
city_sales.plot(kind='area', color='lightgreen', alpha=0.7)
plt.title(f'City-wise Sales in {state}')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
file_path = '/content/drive/MyDrive/samplesuper/Sample - Superstore.xlsx'
df = pd.read_excel(file_path)

# Group by product and calculate total sales, selecting the top 10 products
product_sales = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

# Plotting product-wise total sales as a pie chart with indicators
plt.figure(figsize=(10, 8))

# Colors for the pie chart
colors = sns.color_palette("Set3", len(product_sales))

# Plot the pie chart with indicators
plt.pie(
    product_sales,
    labels=product_sales.index,
    autopct=lambda p: f'{p:.1f}%\n(${p * sum(product_sales) / 100:.0f})',  # Shows percentage and actual sales
    startangle=140,
    colors=colors,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1}
)

# Title and additional customization
plt.title('Top 10 Product-wise Total Sales with Indicators')
plt.legend(product_sales.index, title="Product Name", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.show()

```

```python
# Group by product and calculate quantity sold
product_quantity = df.groupby('Product Name')['Quantity'].sum().sort_values(ascending=False).head(10)

# Plotting product-wise sale quantity as a line chart
plt.figure(figsize=(12, 8))
product_quantity.plot(kind='line', marker='o', color='purple')
plt.title('Top 10 Product-wise Sale Quantity')
plt.xlabel('Product')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=90)
plt.show()
```

```python
# Group by product and calculate profit
product_profit = df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(10)

# Plotting product-wise profit
plt.figure(figsize=(12, 8))
product_profit.plot(kind='bar', color='coral')
plt.title('Top 10 Product-wise Profit')
plt.xlabel('Product')
plt.ylabel('Profit')
plt.xticks(rotation=90)
plt.show()
```

```python
# Calculate total profit
total_profit = df['Profit'].sum()
print(f"Total Profit: ${total_profit:.2f}")
```

```python
# Group by product and calculate average discount
product_discount = df.groupby('Product Name')['Discount'].mean().sort_values(ascending=False).head(10)

# Plotting product-wise discount
plt.figure(figsize=(12, 8))
product_discount.plot(kind='bar', color='gold')
plt.title('Top 10 Product-wise Average Discount')
plt.xlabel('Product')
plt.ylabel('Average Discount')
plt.xticks(rotation=90)
plt.show()
```

```python
# Ensure that the 'Order Date' column is in datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Group by order date and calculate total sales
date_sales = df.groupby('Order Date')['Sales'].sum()

# Plotting date-wise sales
plt.figure(figsize=(14, 8))
date_sales.plot(kind='line', color='blue')
plt.title('Date-wise Sales')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Find the date with the highest sales
highest_sales_date = date_sales.idxmax()
highest_sales_value = date_sales.max()
print(f"The date with the highest sales is {highest_sales_date.date()} with sales amounting to ${highest_sales_value:.2f}")
```

```python
# Group by category and calculate total sales
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

# Plotting category-wise sales
plt.figure(figsize=(10, 6))
category_sales.plot(kind='bar', color='mediumseagreen', edgecolor='black')

# Adding labels and title
plt.title('Category-wise Sales')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)

# Display the bar chart
plt.show()
```
